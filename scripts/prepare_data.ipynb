## Prepare Data

#### Imports

```python
import pandas as pd
import numpy as np
import geopandas as gpd
import shapely
from shapely import Point 
from sklearn.cluster import DBSCAN
from shapely.geometry import box
import os 
```

#### Paths

```python
data_dir = os.path.join(os.path.dirname(os.getcwd()), 'data')

fig_dir = os.path.join(os.path.dirname(os.getcwd()), 'figures')

cerf_data_dir = os.path.join(data_dir, 'cerf_im3_western_us_plant_data', 'power_plant_data')

wecc_ads_fn = os.path.join(data_dir, '2034_ADS_PCM_V3.1(2024LnR)_Public Data/GeneratorList.csv')
```

#### Settings

```python
scenario_list = ['rcp45cooler_ssp3',
                 'rcp45cooler_ssp5',
                 'rcp85cooler_ssp3',
                 'rcp85cooler_ssp5',
                 'rcp45hotter_ssp3',
                 'rcp45hotter_ssp5',
                 'rcp85hotter_ssp3',
                 'rcp85hotter_ssp5'
                ]

color_dict = {
                'Geothermal':'#ad1b0a',
                'Solar':'#F7B615', 
                'Wind':'#95C060',
                'Wind (Offshore)': 'teal',
                'Biomass':'#e30787', 
                'Hydroelectric':'#2B7F9E', 
                'Nuclear': '#813f99', 
                'Petroleum':'#d9c7a9', 
                'Natural Gas':'#f08d02',
                'Coal':'#363636',
}

generation_map_dict={
    'Biomass': 'Biomass',
     'Biomass (CCS)': 'Biomass',
     'Coal': 'Coal',
     'Coal (CCS)': 'Coal',
     'Gas CC': 'Natural Gas',
     'Gas CC (CCS)': 'Natural Gas',
     'Gas CT': 'Natural Gas',
     'Geothermal': 'Geothermal',
     'Hydro': 'Hydroelectric',
     'Nuclear':'Nuclear', 
     'Oil CT':'Petroleum', 
     'Solar CSP':'Solar',
     'Solar PV':'Solar',
     'Wind (Offshore)':'Wind',
     'Wind (Onshore)':'Wind'
}
```

### Collect Power Plant Locations

```python
def combine_scenario_df(cerf_data_dir, scenario_list):
    
    for index, scenario in enumerate(scenario_list):
        
        scenario_df = pd.read_csv(os.path.join(cerf_data_dir, scenario, f'power_plant_data_{scenario}.csv'))

        if index == 0:
            output_df = scenario_df
        else:
            output_df = pd.concat([output_df, scenario_df])

    # assign colors and simplified naming
    output_df['generation_type'] =  output_df['tech_simple'].map(generation_map_dict)
    output_df['color'] = output_df['generation_type'].map(color_dict)

    return output_df

def convert_to_geodataframe(df, target_crs='ESRI:102003'):
    """ Create geometry from x,y coordinates"""
    geometry = [Point(xy) for xy in zip(df['xcoord'], df['ycoord'])]
    gdf = gpd.GeoDataFrame(df, crs=target_crs, geometry=geometry)
    return gdf

df = combine_scenario_df(cerf_data_dir=cerf_data_dir, scenario_list=scenario_list)
gdf = convert_to_geodataframe(df)

# save to file
gdf.to_file(os.path.join(data_dir, 'power_plants_by_sited_year.geojson'))
```

### Organize Timestep Data

Expands dataframe to get operational plants by timestep

```python
col_list = ['timestep', 'region_name', 'generation_type', 'color', 'unit_size_mw', 'scenario']
timestep_list = list(range(2020, 2056, 5))

d = pd.DataFrame()
for s in scenario_list:
    
    d_scenario = gdf[(gdf.scenario == s)].copy()

    for timestep in timestep_list:  
        
        df_timestep = d_scenario[(d_scenario.sited_year <= timestep) & (d_scenario.retirement_year > timestep)].copy()
        
        df_timestep['timestep'] = timestep
        
        df_timestep = df_timestep[col_list]
        
        df_timestep_pivot = pd.pivot_table(df_timestep, values='unit_size_mw', index=['scenario', 'timestep', 'generation_type'],
                              columns=['region_name'], aggfunc='sum').reset_index().rename_axis(None, axis=1).fillna(0)
        
        df_timestep_melt = pd.melt(df_timestep_pivot, id_vars=['scenario', 'timestep', 'generation_type'], 
                                  value_vars=df_timestep_pivot.columns[3:], var_name='region_name', value_name='unit_size_mw')
        
        d = pd.concat([d, df_timestep_melt])

# save to file
d.to_csv(os.path.join(data_dir, 'operational_power_plants_by_timestep.csv'), index=False)
```

### Cluster 2055 Data

```python
timestep = 2055
gdf_non_cerf = gdf[gdf.cerf_sited == False].copy()
gdf_non_cerf = gdf_non_cerf[(gdf_non_cerf.sited_year <= timestep) & (gdf_non_cerf.retirement_year > timestep)]

gdf_non_cerf = gdf_non_cerf[['cerf_plant_id', 'unit_size_mw','scenario','tech_simple','region_name','color', 'geometry']]
gdf_non_cerf = gdf_non_cerf.dissolve(by=['scenario', 'cerf_plant_id'], as_index=False, aggfunc={'unit_size_mw':'sum',
                                                                   'tech_simple':'first',
                                                                   'region_name':'first',
                                                                   'color':'first'})
gdf_non_cerf['geometry'] = gdf_non_cerf.geometry.centroid

tech_list = ['Solar PV', 'Wind (Onshore)', 'Wind (Offshore)']

gdf_cerf = gdf[gdf.cerf_sited == True].copy()
gdf_cerf = gdf_cerf[(gdf_cerf.sited_year <= timestep) & (gdf_cerf.retirement_year > timestep)]

# get non wind/solar
non_tech_df = gdf_cerf[~gdf_cerf.tech_simple.isin(tech_list)]
non_tech_df = non_tech_df[['unit_size_mw','scenario','tech_simple','region_name','color', 'geometry']]

# cluster wind/solar
stacked_gdf = gpd.GeoDataFrame()

for tech in tech_list:
    tech_df = gdf_cerf[gdf_cerf.tech_simple == tech].copy()
    for scenario in scenario_list:
        scenario_df = tech_df[tech_df.scenario == scenario].copy()
        for region in scenario_df.region_name.unique():
            region_df = scenario_df[scenario_df.region_name == region].copy()
            for interconnection_zone in region_df.lmp_zone.unique():
                
                i_data = region_df[region_df.lmp_zone == interconnection_zone].copy()

                for y in i_data.sited_year.unique():
                    d = i_data[i_data.sited_year == y].copy()

                    if d.empty:
                        pass
                    else:
                
                        X = np.array(d[['xcoord','ycoord']], dtype='float64')
                        eps_distance_m = 2000
                        min_generators = 1
            
                        # set up the DBSCAN Model
                        model = DBSCAN(eps=eps_distance_m, min_samples=min_generators).fit(X)
                    
                        # create class labels
                        class_predictions = model.labels_
                    
                        # create a cluster id columns
                        d['cluster_number'] = class_predictions
        
                        data_unclustered = d[d.cluster_number == -1]
                        data_unclustered = data_unclustered[['unit_size_mw','scenario','tech_simple','region_name','color', 'geometry']]
                        
                        data_clustered = d[d.cluster_number != -1]
        
                        data_clustered = data_clustered.dissolve(by='cluster_number', aggfunc={'unit_size_mw':'sum',
                                                                                               'scenario':'first',
                                                                                               'tech_simple':'first',
                                                                                               'region_name':'first',
                                                                                               'color':'first'
                                                                                              },
                                                                 as_index=False)
        
                        data_clustered.geometry = data_clustered.geometry.centroid
        
                        stacked_gdf = pd.concat([stacked_gdf, data_clustered, data_unclustered])

# combine data
data = pd.concat([stacked_gdf, non_tech_df, gdf_non_cerf])

# save to file
data.to_file(os.path.join(data_dir, 'clustered_power_plants_2055.geojson'))
```

```python
ads_generators = pd.read_csv(wecc_ads_fn, skiprows=2, encoding='ISO-8859-1')
ads_generators = ads_generators[['SubType', 'State', 'PSSEMaxCap(MW)']]

tech_bin = {
    'PV-Tracking': 'Solar',
    'Hydro':'Hydroelectric', 
    'CombustionTurbine Gas':'Natural Gas', 
    'Bio':'Biomass',
    'PV-NonTracking': 'Solar',
    'WT-Onshore': 'Wind', 
    'Steam-Gas': 'Natural Gas', 
    'Geothermal':'Geothermal', 
    'Solar CSP': 'Solar',
    'CombinedCycle Gas':'Natural Gas', 
    'Nuclear':'Nuclear', 
    'Coal':'Coal', 
    'PV': 'Solar',
    'ST-NatGas':'Natural Gas', 
    'Geo-BinaryCycle':'Geothermal', 
    'Bio-ST':'Biomass',
    'WT-Offshore': 'Wind', 
    'CT-NatGas-Aero':'Natural Gas'
}

state_list = ['OR', 'CA', 'WA', 'ID', 'MT', 'UT', 'NV', 'WY', 'CO', 'AZ', 'NM']

ads_generators = ads_generators[ads_generators.State.isin(state_list)]
drop_list= [
            'ICE', 
            'HydroRPS',
            'MotorLoad', 
            'DC-Intertie', 
            'DR',
            'Energy Storage', 
            'DG-BTM', 
            'DC Line', 
            'ICE-NatGas', 
            'PumpedStorage']
ads_generators = ads_generators[~ads_generators.SubType.isin(drop_list)]
ads_generators['generation_type'] = ads_generators.SubType.map(tech_bin)
ads_generators = ads_generators[['generation_type', 'State', 'PSSEMaxCap(MW)']].groupby(['generation_type', 'State'], as_index=False).sum()
ads_generators['timestep'] = 2034
ads_generators = ads_generators.rename(columns={'PSSEMaxCap(MW)':'unit_size_mw', 'State':'region_name'})
ads_generators['scenario'] = 'ADS'
ads_generators.generation_type.unique()

ads_generators = ads_generators.pivot(index=['generation_type', 'timestep', 'scenario'], 
                     columns=['region_name'], values='unit_size_mw').reset_index().fillna(0).melt(id_vars=['generation_type', 'timestep', 'scenario'],
                                                                                                 value_name='unit_size_mw')

# add interconnection total
keep_col=['scenario', 'timestep', 'generation_type', 'unit_size_mw']
total = ads_generators[keep_col].groupby(['scenario', 'timestep', 'generation_type'], as_index=False).sum()
total['region_name'] = 'Total'

ads_generators_plot = pd.concat([ads_generators, total])

# save to file
ads_generators_plot.to_csv(os.path.join(data_dir, 'wecc_ads_capacity_2034.csv'), index=False)
```

### Create 50km grid polygon file

```python
def create_grid(bounds, crs="ESRI:102003", cell_size=50000, partial_cells=True):
    minx, miny, maxx, maxy = bounds

    # Generate grid coordinates
    x_coords = np.arange(minx, maxx, cell_size)
    y_coords = np.arange(miny, maxy, cell_size)

    # Include edge if partial_cells=True
    if partial_cells:
        if x_coords[-1] < maxx:
            x_coords = np.append(x_coords, maxx)
        if y_coords[-1] < maxy:
            y_coords = np.append(y_coords, maxy)

    polygons = []
    for x0 in x_coords[:-1]:
        for y0 in y_coords[:-1]:
            x1 = x0 + cell_size
            y1 = y0 + cell_size
            # skip partial cells if they exceed the bounds
            if not partial_cells and (x1 > maxx or y1 > maxy):
                continue
            polygons.append(box(x0, y0, x1, y1))

    gdf = gpd.GeoDataFrame({"geometry": polygons}, crs=crs)
    return gdf

grid_50km = create_grid(bounds=(-2725615.228, -629565.8293000001, -425615.2280000001, 1570434.171))
grid_50km = grid_50km.reset_index(names='grid_id')

grid_50km.to_file(os.path.join(data_dir, 'grid_50km_template.geojson'))
```
